```python
import glob
import librosa
import matplotlib.pyplot as plt
import librosa.display as libdip
import pandas as pd
import numpy as np
import sklearn
```

```python
file_path = glob.glob('/content/drive/MyDrive/Data/testing'+'/*.wav')
audio_data = pd.DataFrame()
```

```python
def mfcc(audio, sr):
  mfcc = librosa.feature.mfcc(audio, n_mfcc=13, sr=sr)
  return mfcc
```

```python
def chroma(audio, sr):
  chroma  = librosa.feature.chroma_cqt(y=audio, sr=sr, fmin=50)
  chroma1 = librosa.feature.chroma_cens(y=audio, sr=sr, fmin=50)
  chroma2 = librosa.feature.chroma_stft(y=audio, sr=sr)
  return chroma,chroma1,chroma2
```

```python
def spectral_centroid(audio, sr):
  spectral_centroid = librosa.feature.spectral_centroid(y=audio, sr=sr)[0]
  return spectral_centroid
```

```python
def scaling(sp):
  return sklearn.preprocessing.minmax_scale(sp)
```

```python
def zcr(audio):
  zero_crossing = librosa.feature.zero_crossing_rate(audio)[0]
  return zero_crossing
```

```python
def rmse(audio):
  rms = librosa.feature.rms(y=audio)[0]
  return rms
```

```python
def spectral_rolloff(audio, sr):
  spectral_rolloff = librosa.feature.spectral_rolloff(audio, sr=sr)[0]
  return spectral_rolloff
```

```python
def chroma_datainput(chroma_data):
  col_name = chroma_data.columns
  #Mean Value
  for i in range(len(col_name)):
    audio_data['Mean_Cens '+str(i)]=np.nan
    audio_data['Mean_Cqt '+str(i)]=np.nan
    audio_data['Mean_Stft '+str(i)]=np.nan
  #Median Value
  for i in range(len(col_name)):
    audio_data['Median_Cens '+str(i)]=np.nan
    audio_data['Median_Cqt '+str(i)]=np.nan
    audio_data['Median_Stft '+str(i)]=np.nan
  #MAX value
  for i in range(len(col_name)):
    audio_data['Max_Cens '+str(i)]=np.nan
    audio_data['Max_Cqt '+str(i)]=np.nan
    audio_data['Max_Stft '+str(i)]=np.nan
  #MIN value
  for i in range(len(col_name)):
    audio_data['Min_Cens '+str(i)]=np.nan
    audio_data['Min_Cqt '+str(i)]=np.nan
    audio_data['Min_Stft '+str(i)]=np.nan
    
  #standard deviation
  for i in range(len(col_name)):
    audio_data['Std_Cens '+str(i)]=np.nan
    audio_data['Std_Cqt '+str(i)] =np.nan
    audio_data['Std_Stft '+str(i)]=np.nan 
  return 
```

```python
def mfcc_datainput(mfcc_data):
  col_name = mfcc_data.columns
  #Mean Value of MFCC
  for i in range(len(col_name)):
    audio_data['Mean_MFCC '+str(i)] = np.nan
  #Median Value of MFCC
  for i in range(len(col_name)):
    audio_data['Median_MFCC '+str(i)] = np.nan
  #MAX value of MFCC
  for i in range(len(col_name)):
    audio_data['Max_MFCC '+str(i)] = np.nan
  #MIN value of MFCC
  for i in range(len(col_name)):
    audio_data['Min_MFCC '+str(i)] = np.nan
  #standdard deviation value of MFCC
  for i in range(len(col_name)):
    audio_data['Std_MFCC  '+str(i)] = np.nan

  return 
```

```python
def appendData(mfcc, cqt, cens, stft,cent,zcr, roll ,rms):
  listData = []
  col_name = mfcc.columns
  col_name_1 = cens.columns
  col_name_2 = cent.columns
  col_name_3 = roll.columns
  col_name_4 = zcr.columns
  col_name_5 = rms.columns
  global count 
  global audio_data
  if count >= 5104:
    listData.append('fake')
  else:
    listData.append('real')
 #MFCC
  for i in range(len(col_name)):
    mean = np.mean(mfcc[col_name[i]])
    listData.append(mean)
  for i in range(len(col_name)):
    median = np.median(mfcc[col_name[i]])
    listData.append(median)
  for i in range(len(col_name)):
    max = np.max(mfcc[col_name[i]])
    listData.append(max)
  for i in range(len(col_name)):
    min = np.min(mfcc[col_name[i]])
    listData.append(min)
  for i in range(len(col_name)):
    std = np.std(mfcc[col_name[i]])
    listData.append(std)

 #Chroma 
  for i in range(len(col_name_1)):
    mean0 = np.mean(cens[col_name_1[i]])
    mean1 = np.mean(cqt[col_name_1[i]])
    mean2 = np.mean(stft[col_name_1[i]])
    listData.append(mean0)
    listData.append(mean1)
    listData.append(mean2)

  for i in range(len(col_name_1)):
    median0 = np.median(cens[col_name_1[i]])
    median1 = np.median(cqt[col_name_1[i]])
    median2 = np.median(stft[col_name_1[i]])
    listData.append(median0)
    listData.append(median1)
    listData.append(median2)

  for i in range(len(col_name_1)):
    max0 = np.max(cens[col_name_1[i]])
    max1 = np.max(cqt[col_name_1[i]])
    max2 = np.max(stft[col_name_1[i]])
    listData.append(max0)
    listData.append(max1)
    listData.append(max2)

  for i in range(len(col_name_1)):
    min0 = np.min(cens[col_name_1[i]])
    min1 = np.min(cqt[col_name_1[i]])
    min2 = np.min(stft[col_name_1[i]])
    listData.append(min0)
    listData.append(min1)
    listData.append(min2)

  for i in range(len(col_name_1)):
    std0 = np.std(cens[col_name_1[i]])
    std1 = np.std(cqt[col_name_1[i]])
    std2 = np.std(stft[col_name_1[i]])
    listData.append(std0)
    listData.append(std1)
    listData.append(std2)
#Centroid

  for i in range(len(col_name_2)):
    mean = np.mean(cent[col_name_2[i]])
    listData.append(mean)
  for i in range(len(col_name_2)):
    median = np.median(cent[col_name_2[i]])
    listData.append(median)
  for i in range(len(col_name_2)):
    max = np.max(cent[col_name_2[i]])
    listData.append(max)
  for i in range(len(col_name_2)):
    min = np.min(cent[col_name_2[i]])
    listData.append(min)
  for i in range(len(col_name_2)):
    std = np.std(cent[col_name_2[i]])
    listData.append(std)
#Rolloff

  for i in range(len(col_name_3)):
    mean = np.mean(roll[col_name_3[i]])
    listData.append(mean)
  for i in range(len(col_name_3)):
    median = np.median(roll[col_name_3[i]])
    listData.append(median)
  for i in range(len(col_name_3)):
    max = np.max(roll[col_name_3[i]])
    listData.append(max)
  for i in range(len(col_name_3)):
    min = np.min(roll[col_name_3[i]])
    listData.append(min)
  for i in range(len(col_name_3)):
    std = np.std(roll[col_name_3[i]])
    listData.append(std)
#ZCR

  for i in range(len(col_name_4)):
    mean = np.mean(zcr[col_name_4[i]])
    listData.append(mean)
  for i in range(len(col_name_4)):
    median = np.median(zcr[col_name_4[i]])
    listData.append(median)
  for i in range(len(col_name_4)):
    max = np.max(zcr[col_name_4[i]])
    listData.append(max)
  for i in range(len(col_name_4)):
    min = np.min(zcr[col_name_4[i]])
    listData.append(min)
  for i in range(len(col_name_4)):
    std = np.std(zcr[col_name_4[i]])
    listData.append(std)

#RMS  
  for i in range(len(col_name_5)):
    mean = np.mean(rms[col_name_5[i]])
    listData.append(mean)
  for i in range(len(col_name_5)):
    median = np.median(rms[col_name_5[i]])
    listData.append(median)
  for i in range(len(col_name_5)):
    max = np.max(rms[col_name_5[i]])
    listData.append(max)
  for i in range(len(col_name_5)):
    min = np.min(rms[col_name_5[i]])
    listData.append(min)
  for i in range(len(col_name_5)):
    std = np.std(rms[col_name_5[i]])
    listData.append(std)

  audio_data.loc[count] = listData
  count+=1
  return 
```
